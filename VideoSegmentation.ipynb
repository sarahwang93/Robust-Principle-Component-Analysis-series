```python
def conv_diagnal(singular_value, rank):
    s_diag = np.zeros((rank, rank))
    for i in range(0, rank):
        for j in range(0, rank):
                if i == j:
                    s_diag[i][j] = singular_value[i]
    return s_diag           
```

```python
#algorithm goes here 
import numpy as np 
import matplotlib 
import matplotlib.pyplot as plt 
from timeit import timeit
from PIL import Image
# from skimage import color,io
import scipy
from sklearn.decomposition import TruncatedSVD
import random
import cv2
import torch
import torchvision.transforms.functional as fn
#from plotter import convergence_plot, kwargs, setup_layout

def sp_noise(image,prob):
    '''
    Add salt and pepper noise to image
    prob: Probability of the noise
    '''
    output = np.zeros(image.shape,np.uint8)
    thres = 1 - prob 
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            rdn = random.random()
            if rdn < prob:
                output[i][j] = 0
            elif rdn > thres:
                output[i][j] = 255
            else:
                output[i][j] = image[i][j]
    return output


def convert_RGB(input_img):
    if input_img.ndim == 2 or (input_img.ndim==3 and input_img.shape[2] ==1):
        # your image is in grayscale
        input_img = color.gray2rgb(input_img)
    elif input_img.ndim==3 and input_img.shape[2] == 4:
        # your image is in RGBA
        input_img = color.rgba2rgb(input_img)
    else:
        # your image is already in RGB
        assert input_img.ndim == 3 and input_img.shape[2] == 3
    return input_img
```

```python
import math
from skimage import io

shape1 = 240 #1080
shape2 = 320 #1920
shape3 = 1 
shape_0 = (240, 320)    
n_1 = max(shape1, shape2)
n_2 = min(shape1, shape2)
lam = 1/np.sqrt(n_1)
m = 0.1 * n_1 * n_2
mu = (shape1*shape2)/(4 * np.linalg.norm(imgGray[0], 1, keepdims=True))

gamma = 0.05
beta = 1.618
rho = 0.01
k = 0.01 * n_1 * n_1
lam_bda = 10

img = io.imread('./survelliance/0034.jpg')
img_res = convert_RGB(img)
print(img.shape)

#input
imgGray = color.rgb2gray(img_res)
print(imgGray*255)
shape1 = imgGray.shape[0]
shape2 = imgGray.shape[1]

D = imgGray
S_sparse = np.zeros((shape1, shape2))#


t = np.zeros((shape1, shape2))
D = imgGray
# k is the iteration number
rho_k = rho * k

pi = np.zeros((shape1, shape2))
S  = np.zeros((shape1, shape2))


# Lagrangian multiplier
# size of pi is m*n
# hereby S is representing sparse error data  which is basing on the decomposition result
# S is the sparse matrix 
def lagrangian_mul(U, V, S, pi, rho, iter_num, rank):
        pi = D * rho * (iter_num-1)
        # original paper has a differente version with this one 
        # lagrangian = nuclear_norm(low_rank) + lam_bda * numpy.linalg.norm(sparse_tensor, ord=1) + np.inner(pi, D -  np.matmul(U, V) - S) + mu/2 * numpy.linalg.norm(D - np.matmul(U, V)-sparse,ord='fro', keepdims=True) ** 2
        res_1 = np.linalg.norm(S, ord=1) 
        res_2 = lam_bda * subgradient(V, rank, gamma)  
        res_3 = np.inner(D * rho * (iter_num-1), D - np.matmul(U, V) - S)
        res_4 = rho/2 * np.linalg.norm(D -  np.matmul(U, V) - S, ord='fro')**2
        return res_1 + res_2 ,res_3 + res_4
         
# u is constant to get next u 
# u is 1080 * r, vofT is r * 1920
# T and S are 1080*1920

# update steps for u
def orthogonal_proc(v, s, iter_num, rank_of_orth):
    pi = D * rho * (iter_num-1)
    t = D + pi/rho_k
    if v.shape == (shape2, 5):
        pass
    else:
        v = np.transpose(v)  
    left, singular, right = svd_res(np.matmul(t-s, v), rank_of_orth)
    return np.matmul(left, right)


def solve_u(u_input, v, s):
    t = D + pi/rho_k
    u = u_input
    u = rho_k/2 * numpy.linalg.norm(np.matmul(u, v)-(t-s), ord="fro") ** 2
    return u


# update for v
# gamma is larger than 0 
def subgradient(v, rank, gamma):
    singular_proc = []
    left, singular, right = svd_res(v, rank)
    for i in singular:
        each_sin = math.exp(-i/gamma)/gamma 
        singular_proc.append(each_sin)
    diag_ma = conv_diagnal(singular_proc, rank)
    return np.matmul(np.matmul(left, diag_ma), right)

def control_ratio(v,gamma):
    sum = 0
    singular_value = scipy.linalg.svdvals(v)
    num = len(singular_value)
    for i in range(0, num):
        sum += (1-math.exp(-singular_value[i]/gamma))
    return sum

# v is a constant 
# return updated v
def solve_v(v_input, u, s):
    v = v_input
    v_new = control_ratio(v, gamma) + (rho_k/2*lam)*numpy.linalg.norm(np.transpose(v) - np.matmul(np.transpose(t - s), u), ord="fro") ** 2
    return v_new

def optimal_res_v(v_input, sparse_input, u_updated):
    t = D + pi/rho_k 
    v_1 = np.matmul(np.transpose(t-sparse_input),u_updated)  
    v_2 = lam_bda * subgradient(v_input, 3, gamma)/rho_k
    if v_2.shape == (shape2, 5):
        pass
    else:
        v_2 = np.transpose(v_2)  
    v_updated = v_1 - v_2
    return v_updated
    

# update steps for S
def shrinkage_thre(u, v):
    t = D + pi/rho_k
    W= t - numpy.matmul(u,np.transpose(v))
    update_S = np.zeros((t.shape[0], t.shape[1]))
    for i in range(0, len(W)):
        for j in range(0, len(W[0])):
            res = max(abs(numpy.matrix.max(np.matrix(W[i][j]))) - 1/rho_k, 0) * numpy.sign(W[i][j])
            update_S[i][j] = res
    return update_S

# s is a constant 
def solve_s(sparse_input, u, v):
    s = sparse_input
    s_new = numpy.linalg.norm(s, ord=1) + rho_k/2 * numpy.linalg.norm(s-(t-np.matmul(u,v)),ord="fro") ** 2
    return s_new
```
